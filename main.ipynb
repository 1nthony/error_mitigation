# Comparison ZNE: circuit folding

```python
dic_to_compare = {
    'exp': 'ZNE_exp_circuit', 
    'lin': 'ZNE_lin_circuit', 
    'quad': 'ZNE_quad_circuit', 
    'rich_d3': 'ZNE_rich_circuit'
}
```

```python
n_qubits = 3

Nshots = 1000
noise_levels = np.linspace(1, 200, 70)
noise_levels = noise_levels[np.arange(0, len(noise_levels), 4)]
nreps = 150
possible_outputs = generate_binary_values(n_qubits)

depolarizing_noise = 0.01
noise_type = 'depolarizing'  # 'fake_perth', 'depolarizing'
params_noise = [depolarizing_noise]


run_existing_circuits = True
for zne_type,sim_folder in dic_to_compare.items() :
        folding = lambda x: circuit_folding(x, num_bits=n_qubits, sim_folder=sim_folder, circuits_exist=run_existing_circuits)
        doit(   sim_folder, zne_type, nreps, folding, possible_outputs, 
                Nshots, noise_type, noise_levels, *params_noise)
        # run_existing_circuits = True
```
